## Anonymizing winner and loser 
In an attempt to remove bias from the dataframe, I'm anonymizing the winners and losers for each matches. First, I'm shuffling the values between the winner and loser columns and renaming each column player 1 and player 2. 

```python
#separating the winner, loser and neutral columns
winner_cols = [col for col in ausopen.columns if col.startswith('w')]
loser_cols = [col for col in ausopen.columns if col.startswith('l')]
other_cols = ['tourney_id','match_num','score','round','minutes']
```

```python
ausopen.winner_name.value_counts()
```

```python
#dividing the dataframe randoming in two halves 

half_ausopen = ausopen.sample(frac=0.5, random_state=1)
```

```python
half_ausopen.shape
```

```python
other_half = ausopen.drop(half_ausopen.index)
```

```python
other_half.shape
```

```python
ausopen.shape
```

```python
##for one half the winner will be player 1 
#for the other half the winner will be player 2 
winner_fhalf = [col for col in half_ausopen.columns if col.startswith('w')]
loser_fhalf = [col for col in half_ausopen.columns if col.startswith('l')]
other_fhalf = ['tourney_id','match_num','score','round','minutes', 'year']
```

```python
fhalf = half_ausopen[winner_fhalf + loser_fhalf + other_fhalf]
```

```python
new_column_names_w = [col.replace('w_','winner_') for col in winner_fhalf]
new_column_names_l = [col.replace('l_','loser_') for col in loser_fhalf]
```

```python
final_col_names_w = [col.replace('winner_','p2') for col in new_column_names_w]
final_col_names_l = [col.replace('loser_','p1') for col in new_column_names_l]
final_col_names_l
```

```python
fhalf.columns = final_col_names_w + final_col_names_l + other_fhalf
```

```python
fhalf.head()
```

```python
#creating a column to denote whether player 1 or player 2 won 
fhalf['p1won'] = 0
fhalf['p2won'] = 1
```

```python
fhalf.head()
```

```python
##other_half - winner player 1 
winner_shalf = [col for col in other_half.columns if col.startswith('w')]
loser_shalf = [col for col in other_half.columns if col.startswith('l')]
other_shalf = ['tourney_id','match_num','score','round','minutes', 'year']
```

```python
shalf = other_half[winner_shalf + loser_shalf + other_shalf]
```

```python
new_column_names_sw = [col.replace('w_','winner_') for col in winner_shalf]
new_column_names_sl = [col.replace('l_','loser_') for col in loser_shalf]
```

```python
final_col_names_sw = [col.replace('winner_','p1') for col in new_column_names_w]
final_col_names_sl = [col.replace('loser_','p2') for col in new_column_names_l]
final_col_names_sl
```

```python
shalf.columns =  final_col_names_sl + final_col_names_sw + other_shalf
```

```python
##re-do this 

shalf['p1won'] = 1
shalf['p2won'] = 0
```

## Analyzing metrics for possible model features

```python
#plotting correlation graph
plt.figure(figsize=(15,15))
final_corr = final_1.corr()

sns.heatmap(final_corr, vmin=-1, vmax=1, annot=True, cmap='coolwarm');
```

### Observations:
- **Player two age:** Fairly negative correlation between player one's win and player two's age. Therefore, player one is more likely to win if player two is older. 
- **Number of aces (either player):** Number of aces player one scores is positively correlated to player one's win. Thus, the higher the number of aces the higher the possibility of a win. 
- **Rank (either player):** Higher rank (e.g #1, #2, etc..) negatively correlated to a win. Therefore the higher a player ranks, the higher his probability to win the match. 

```python
#checking if higher ranked players tend to win 
plt.figure(figsize=(6,6))
sns.scatterplot(x=final_1['p1rank'], y=final_1['p1won']);
```

```python
#checking the same for player 2 
plt.figure(figsize=(10,10))
sns.scatterplot(x=final_1['p2rank'], y=final_1['p2won'])
```

```python
#checking for instances where the player who won ranked higher than the player who lost 

final_1[((final_1.p1won == 1) & (final_1.p2rank < final_1.p1rank)) |((final_1.p2won == 1) & (final_1.p1rank < final_1.p2rank))]
```

```python
1-(610/1270)
```

There's been 610 instances out of 1270 where the player who won ranked lower than the player who lost. Thus, rank has been an accurate predictor of winner about 52% of the time.
