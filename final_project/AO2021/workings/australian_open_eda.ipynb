## Summary

### Shortcomings: 
- The scores for each match are not displayed in the most usable manner. The scores for all sets in a match are lumped and listed together hence any statistic derived from the score column is likely flawed and misleading.
- Each year’s selection of players includes new, unranked players. Therefore, it will be hard to get head to head statistics for those players and hard to gauge how well the player will fare against any given player.

### Assumptions: 
- Some players did not have a rank in their earlier matches of their career. I looked up the earliest available rank for them at the time of the match and assigned that to them. 
- The draw of players for the 2021 Australian Open will happen sometime in February (it has unfortunately been postponed by a month) so I’m assuming that the same players that were picked in 2020 will play in 2021.
