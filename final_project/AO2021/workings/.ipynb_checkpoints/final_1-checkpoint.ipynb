```python
import pandas as pd
from matplotlib import pyplot as plt
from datetime import datetime
import seaborn as sns

%matplotlib inline
```

```python
wins = pd.read_csv('./data/wins.csv')
wins_hsf = pd.read_csv('./data/win_hard.csv')
```

**Data Dictionary**

- `player`: Player name
- `player_id`: Player ID  
- `rank`: Current rank of player
- `rank_points`: Current number of rank points player has
- `wins`: Total number of wins in the past 3 years 
- `losses`: Total number of losses in the past 3 years 
- `matches` : Total matches played in the past 3 years
- `win_pct`: Percentage wins in the past 3 years 
- `wins_hsf`: Total number of wins on hard surfaces in the past 3 years 
- `losses_hsf`: Total number of losses on hard surfaces in the past 3 years
- `matches_hsf` : Total number of matches playedd on hard surfaces in the past 3 years 
- `win_pct_hsf` : Percentage wins on hard surfaces in the past 3 years
